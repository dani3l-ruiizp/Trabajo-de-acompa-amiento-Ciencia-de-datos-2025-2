## Limpieza de datos

Se limpia la base de datos, omitiendo variables no relevamtes o sin información con el fin de evitar interferencias en el analisis.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/REGISTROS_SANITARIOS_DE_MEDICAMENTOS_HOMEOPÁTICOS_20251219.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Se evidencia la presencia de valores nulos únicamente en algunas variables, principalmente asociadas a fechas.
df.isnull().sum()
```

```python
df = df.drop(columns=['FECHA_EXPEDICION', 'FECHA_VENCIMIENTO', "MARCA", "CONTRA INDICACIONES Y ADVERTENCIAS"])

#Se eliminan las valores nulos
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()

#Se eliminan registros duplicados
```

```python
df = df.drop_duplicates()
```

```python
columnas_texto = [
    'PRODUCTO', 'TIPOPRODUCTO', 'TITULAR',
    'CIUDAD_TITULAR', 'MODALIDAD', 'ESTADO_REGISTRO'
]

for col in columnas_texto:
    if col in df.columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.strip()
            .str.upper()
        )
```

```python
df.head

#Se comprueba que la cantidad de datos este acorde con los requerimientos solicitados
```

```python
df.info
```

Durante la etapa de limpieza de datos se realizó la verificación de valores nulos, la eliminación de columnas completamente vacías y la depuración de registros duplicados. Adicionalmente, se estandarizaron variables categóricas sin realizar, con el fin de preservar la información original del conjunto de datos.
